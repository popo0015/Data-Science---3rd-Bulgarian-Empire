```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data/w2c3.csv")
```

Spelling Errors

```python
df["City"].unique()
```

```python
df["City"] = df["City"].str.replace("Nwe York", "New York")
```

```python
df["City"].unique()
```

Duplications

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
```

```python
df[df.duplicated()]
```

```python
df.reset_index(drop=True, inplace=False)
```

Missing values

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
plt.hist(df["Income"], bins=30)
plt.show()
```

```python
df["Income"] = df["Income"].replace(-999.0, np.nan)
```

```python
df = df.dropna()
```

```python
plt.hist(df["Income"])
plt.title('Income plot without empty values')
plt.ylabel('count')
plt.xlabel('income in euros')
plt.show()
```

```python
df["Income"] = df["Income"].fillna(df["Income"].median())
```

```python
plt.hist(df["Income"], bins=30)
plt.show()
```
